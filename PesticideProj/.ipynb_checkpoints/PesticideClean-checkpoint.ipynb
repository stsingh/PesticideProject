```python
import numpy as np
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import Lipinski
import pubchempy as pcp
```

```python
df = pd.read_csv("FullPesticide.csv")
df.head()
```

```python
df.drop(["Sample ID", "Commod", "Pesticide Code", "Test Class", "Concentration", "pp_", "Confirm 1", "Confirm 2", "Annotate","Quantitate", "Mean", "Extract", "Determ"], inplace=True, axis=1)
df.drop_duplicates(subset ="Pesticide Name",
                     keep = 'first', inplace = True)
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["EPA Tolerance (ppm)"], inplace=True)
df = df.reset_index(drop=True)
df
```

```python
from urllib.request import urlopen
from urllib.parse import quote

def CIRconvert(ids):
    try:
        url = 'http://cactus.nci.nih.gov/chemical/structure/' + quote(ids) + '/smiles'
        ans = urlopen(url).read().decode('utf8')
        return ans
    except:
        return 'Did not work'
identifiers = df["Pesticide Name"]
smiles = []
for ids in identifiers :
    smiles.append(CIRconvert(ids))
```

```python
np.vstack(smiles);
df['Smiles'] = smiles
df = df[df["Smiles"] != "Did not work"]
df = df.reset_index(drop=True)
```

```python
smiles = df["Smiles"]
df.drop("Smiles", inplace = True, axis = 1)
mols = []
for sm in smiles :
  mols.append(Chem.MolFromSmiles(sm))
```

```python
np.vstack(mols)
df['Molecules'] = mols
df.replace("", nan_value, inplace=True)
df.dropna(subset = ["Molecules"], inplace=True)
mols = df['Molecules']
df.drop("Molecules", inplace = True, axis = 1)
```

```python
molwts = []
hvmolwts = []
numrads = []
numvals = []
hvatoms = []
nhoh = []
no = []
aliccyc = []
alihcyc = []
alirings = []
aroccyc = []
arohcyc = []
arorings = []
haccepts = []
hdonors = []
hetatoms = []
rotbonds = []
satccyc = []
sathcyc = []
satrings = []
rings = []
for mol in mols :
  molwts.append(Chem.Descriptors.ExactMolWt(mol))
  hvmolwts.append(Chem.Descriptors.HeavyAtomMolWt(mol))
  numrads.append(Chem.Descriptors.NumRadicalElectrons(mol))
  numvals.append(Chem.Descriptors.NumValenceElectrons(mol))
  hvatoms.append(Chem.Lipinski.HeavyAtomCount(mol))
  nhoh.append(Chem.Lipinski.NHOHCount(mol))
  no.append(Chem.Lipinski.NOCount(mol))
  aliccyc.append(Chem.Lipinski.NumAliphaticCarbocycles(mol))
  alihcyc.append(Chem.Lipinski.NumAliphaticHeterocycles(mol))
  alirings.append(Chem.Lipinski.NumAliphaticRings(mol))
  aroccyc.append(Chem.Lipinski.NumAromaticCarbocycles(mol))
  arohcyc.append(Chem.Lipinski.NumAromaticHeterocycles(mol))
  arorings.append(Chem.Lipinski.NumAromaticRings(mol))
  haccepts.append(Chem.Lipinski.NumHAcceptors(mol))
  hdonors.append(Chem.Lipinski.NumHDonors(mol))
  hetatoms.append(Chem.Lipinski.NumHeteroatoms(mol))
  rotbonds.append(Chem.Lipinski.NumRotatableBonds(mol))
  satccyc.append(Chem.Lipinski.NumSaturatedCarbocycles(mol))
  sathcyc.append(Chem.Lipinski.NumSaturatedHeterocycles(mol))
  satrings.append(Chem.Lipinski.NumSaturatedRings(mol))
  rings.append(Chem.Lipinski.RingCount(mol))
```

```python
df = df.reset_index(drop=True)
df['ExactMolWt'] = molwts
df['HeavyAtomMolWt'] = hvmolwts
df['NumRadicalElectrons'] = numrads
df['NumValenceElectrons'] = numvals
df['HeavyAtomCount'] = hvatoms
df['NHOHCount'] = nhoh
df['NOCount'] = no
df['NumAliphaticCarbocycles'] = aliccyc
df['NumAliphaticHeterocycles'] = alihcyc
df['NumAliphaticRings'] = alirings
df['NumAromaticCarbocycles'] = aroccyc
df['NumAromaticHeterocycles'] = arohcyc
df['NumAromaticRings'] = arorings
df['NumHAcceptors'] = haccepts
df['NumHDonors'] = hdonors
df['NumHeteroatoms'] = hetatoms
df['NumRotatableBonds'] = rotbonds
df['NumSaturatedCarbocycles'] = satccyc
df['NumSaturatedHeterocycles'] = sathcyc
df['NumSaturatedRings'] = satrings
df['RingCount'] = rings
```

```python
df.to_csv(r"/content/drive/MyDrive/PesticideProj/Data/PesticideDataFullNums.csv", index=False)
```

```python
df["EPA Tolerance (ppm)"] = df["EPA Tolerance (ppm)"].astype(str)
df["EPA Tolerance (ppm)"] = df['EPA Tolerance (ppm)'].str.replace('[^\w\s]','')
df["EPA Tolerance (ppm)"] = df['EPA Tolerance (ppm)'].str.replace('0','')
for num in df["EPA Tolerance (ppm)"] :
  if num.isdigit() :
    df.replace(num, "T", inplace=True)
```
